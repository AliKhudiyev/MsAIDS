```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Returns database
# db format: [SID_0, SID_1, ...]
# SID format: {t0: PATTERN_0, t1: PATTERN_1}
# PATTERN format: [e1, e2, ...]
'''
i.e. db=[
            {
                t00: [e000, e001, ...], 
                t01: [e010, e011, ...], 
                ...
            }, 
            {
                t10: [e100, e101, ...], 
                t11: [e110, e111, ...], 
                ...
            }, 
            ...
        ]
'''
def parse(filepath):
    db = []
    with open(filepath, 'r') as f:
        lines = f.readlines()
    f.close()
    
    for line in lines:
        patterns = line.split(',')
        patterns[-1] = patterns[-1][:-1]
        sequence = {}
        for pattern in patterns:
            index = pattern.find(' ')
            timestamp = int(pattern[1:index])
            event = int(pattern[index+1:len(pattern)-1])
            if timestamp not in sequence:
                sequence[timestamp] = []
            sequence[timestamp].append(event)
                
        db.append(sequence)
    return db

# Returns all timestamps corresponding to event a
def get_all_timestamps(db, a):
    timestamps = []
    for sequence in db:
        for timestamp, pattern in sequence.items():
            for event in pattern:
                if a == event:
                    timestamps.append(timestamp)
                    break
    return timestamps

# Returns all collections which contain A -> (?,b) where A = (t,a)
# Output format: [t0, t1, ...]
def collect(db, A, b):
    timestamps = []
    for sequence in db:
        ts = list(sequence.keys())
        for i, timestamp in enumerate(ts):
            if A[0] == timestamp:
                for t in ts[i+1:]:
                    if b in sequence[t]:
                        timestamps.append(t)
                        break
    
    return timestamps

# Returns all collections which contain (?,a) -> (?,b)
# Output format: [{t0: [t00, t01, ...]}, {t1: [t10, t11, ...]}, ...]
def collect_all(db, a, b):
    X = get_all_timestamps(db, a)
    Y = []
    for i, x in enumerate(X):
        if i and i % 1000 == 0:
            print(f'finished {int(i/1000)}K')
        if i == 3000:
            break
        Y.append(collect(db, (x, a), b))
    return X, Y
```

```python
db = parse('sequences.csv')
X, Y = collect_all(db, -1, -1)
```

```python
plt.xlabel('t1')
plt.ylabel('t2')
delta_T = []
for i in range(0, 1000):
    delta_T.append([y-X[i] for y in Y[i]])
    plt.scatter([X[i]]*len(Y[i]), Y[i]);
```

```python
plt.xlabel('timestamp')
plt.ylabel('delta t')
for i in range(0, 1000):
    plt.scatter([X[i]]*len(Y[i]), delta_T[i]);
```

```python
plt.xlabel('timestamp')
plt.ylabel('delta t')
for i in range(1000):
    if len(delta_T[i]):
        plt.scatter(X[i], np.mean(delta_T[i]));
```

```python
X_Y = []
for i in range(1000):
    if len(delta_T[i]):
        X_Y.append([X[i], np.mean(delta_T[i])])

fig, axs = plt.subplots(1, 5, figsize=(25, 5))
for degree in range(2, 7):
    poly = np.poly1d(np.polyfit(np.array(X_Y)[:,0], np.array(X_Y)[:,1], degree))
    xs = np.linspace(0, 1000, 1000)
    axs[degree-2].plot(xs, poly(xs), c='r');
    axs[degree-2].set_title(f'poly-{degree}')
    
    for i in range(1000):
        axs[degree-2].scatter(np.array(X_Y)[:,0], np.array(X_Y)[:,1], c='#1f77b4');
```

```python
def predict(timestamp):
    result = round(poly([timestamp])[0])
    print(f'After the event -1 occuring at timestamp {timestamp}, upcoming event -1 will occur approximately after {result} time units.')

predict(905)
```
