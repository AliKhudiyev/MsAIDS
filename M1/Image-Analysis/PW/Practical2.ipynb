## I - Histogram visualisation

```python
def pdf(M):
    pixels = M.flatten(order='C')
    unique_pixels = np.unique(pixels)
    frequencies = np.zeros(shape=(len(unique_pixels)))
    for pixel in pixels:
        frequencies[np.where(unique_pixels == pixel)[0]] += 1
    return unique_pixels, frequencies/len(pixels)
```

```python
M = np.zeros(shape=(6,6))
for i in range(6):
    for j in range(6):
        M[i][j] = 4*i+j
print(M)
uniq, freq = pdf(M)

print(uniq)
print(freq)
print(freq*len(M.flatten(order='C')))
```

## II - Application on real images

```python
def hist(imgs):
    fig, axs = plt.subplots(1,len(imgs),figsize=(6*len(imgs),6))
    
    for i, img in enumerate(imgs):
        unique_pixels, frequencies = pdf(img)
        if len(imgs) > 1:
            axs[i].plot(unique_pixels, frequencies);
        else:
            axs.plot(unique_pixels, frequencies);

def imshow(imgs):
    fig, axs = plt.subplots(1,len(imgs),figsize=(6*len(imgs),6))
    
    for i, img in enumerate(imgs):
        if len(imgs) > 1:
            axs[i].imshow(img, 'gray');
        else:
            axs.imshow(img, 'gray');
```

```python
def cumulated_hist(imgs):
    fig, axs = plt.subplots(1,len(imgs),figsize=(6*len(imgs),6))
    
    for i, img in enumerate(imgs):
        unique_pixels, frequencies = pdf(img)
        for j in range(1,len(frequencies)):
            frequencies[j] += frequencies[j-1];
        if len(imgs) > 1:
            axs[i].plot(unique_pixels, frequencies);
        else:
            axs.plot(unique_pixels, frequencies);
```

```python
cumulated_hist([spine, rice, cameraman])
```

## III - Contrast enhancement

### 1. Image binarisation

```python
def binim(img,s):
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            img[i][j] = 0 if img[i][j] < s else 255
```

```python
spine_binary = copy.deepcopy(spine)
binim(spine_binary, 20)

rice_binary = copy.deepcopy(rice)
binim(rice_binary, 100)

cameraman_binary = copy.deepcopy(cameraman)
binim(cameraman_binary, 150)

imshow([spine_binary, rice_binary, cameraman_binary])
cumulated_hist([spine_binary, rice_binary, cameraman_binary])
```

**Comments:** The threshold for each image can easily be obtained from the cumulative distribution function. When cdf(x)=0.5 it means that the gray pixels from 0 to x has 50% percent of the image covered while the pixels from x+1 to 255 are the other half of the image. That's why choosing such pixel values(i.e. 100 for rice) results new cdf in a way that cdf(0)=0.5 and cdf(255)=1 in an ideal case.

### 2. Linear stretching

```python
def recdyn(img,a,b):
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            img[i][j] = a*img[i][j]+b
```

```python
spine_stretched = copy.deepcopy(spine)
recdyn(spine_stretched,50,140)

rice_stretched = copy.deepcopy(rice)
recdyn(rice_stretched,50,140)

cameraman_stretched = copy.deepcopy(cameraman)
recdyn(cameraman_stretched,50,140)

imshow([spine, rice, cameraman])
hist([spine, rice, cameraman])
imshow([spine_stretched, rice_stretched, cameraman_stretched])
hist([spine_stretched, rice_stretched, cameraman_stretched])
```

**Comments:** The first obtained image is not bad in comparison with the second one and the second one is better then the third one which almost looks like a random noise.

### 3. Negative of an image

```python
def negim(img):
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            img[i][j] = 255 - img[i][j]
```

```python
spine_negative = copy.deepcopy(spine)
negim(spine_negative)

rice_negative = copy.deepcopy(rice)
negim(rice_negative)

cameraman_negative = copy.deepcopy(cameraman)
negim(cameraman_negative)

imshow([spine, rice, cameraman])
hist([spine, rice, cameraman])
imshow([spine_negative, rice_negative, cameraman_negative])
hist([spine_negative, rice_negative, cameraman_negative])
```

**Comments:** Since the negative of an image is obtained by setting each pixel value to "255 minus that pixel value" it is obvious that pdf_old(x)=pdf_new(255-x). That is the reason of the vertical mirror effect on the pdf plots.

### 4. Contraction of dark areas and dilatation of bright ones

```python
def t(f,a,b):
    return b/a*f if (0 <= f and f <= a) else ((255-b)*f+255*(b-a))/(255-a)

def dilcont(img,a,b,pc=0,pd=0):
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            img[i][j] = t(img[i][j],a,b)
```

```python
# Brightened images
spine_brightened = copy.deepcopy(spine)
dilcont(spine_brightened,20,100)

rice_brightened = copy.deepcopy(rice)
dilcont(rice_brightened,100,150)

cameraman_brightened = copy.deepcopy(cameraman)
dilcont(cameraman_brightened,150,225)

# Darkened images
spine_darkened = copy.deepcopy(spine)
dilcont(spine_darkened,100,20)

rice_darkened = copy.deepcopy(rice)
dilcont(rice_darkened,150,100)

cameraman_darkened = copy.deepcopy(cameraman)
dilcont(cameraman_darkened,225,150)
```

```python
imshow([spine,rice,cameraman])
hist([spine,rice,cameraman])
cumulated_hist([spine,rice,cameraman])
```

```python
imshow([spine_brightened,rice_brightened,cameraman_brightened])
hist([spine_brightened,rice_brightened,cameraman_brightened])
cumulated_hist([spine_brightened,rice_brightened,cameraman_brightened])
```

```python
imshow([spine_darkened,rice_darkened,cameraman_darkened])
hist([spine_darkened,rice_darkened,cameraman_darkened])
cumulated_hist([spine_darkened,rice_darkened,cameraman_darkened])
```

**Comments:** Brightened images have the pdf and cdf plots strected while darkened images have the pdf and cdf plots shrinked. The reason is that there should be more brighter pixels in the brightened images(which obviously cause brighter pixels appear more and therefore, pdf shows higher values for brighter pixels) and more darker pixels in the darkened images(which cause pdf to show higher values for darker pixels).
