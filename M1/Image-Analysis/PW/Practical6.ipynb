```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import copy
from sklearn.cluster import KMeans
```

```python
lena_gray = cv2.imread('base-images/lena_gray.tif', 0)
plt.imshow(lena_gray, 'gray');
```

### Utilities

```python
# Display the given image with a threshold
# If pixel is smaller than threshold then it's black,
# Otherwise it's white
def display(img, threshold):
    img_copy = copy.deepcopy(img)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if img[i][j] < threshold:
                img_copy[i][j] = 0
            else:
                img_copy[i][j] = 255
    plt.imshow(img_copy, 'gray');

# Probability of a pixel in the image according to its intensity
def pixel_probability(img, intensity, greater):
    pixels = np.unique(img.flatten(order='C'))
    size = len(img.flatten(order='C'))
    
    probabilities = []
    
    if greater:
        for pixel in pixels:
            probability = len(img[img == pixel]) / size
            if probability > intensity:
                probabilities.append(probability)
        return np.sum(probabilities)
    else:
        for pixel in pixels:
            probability = len(img[img == pixel]) / size
            if probability <= intensity:
                probabilities.append(probability)
        return np.sum(probabilities)

# Mean of all pixels in the image according to its intensity
def pixel_mean(img, intensity, greater):
    pixels = np.unique(img.flatten(order='C'))
    size = len(img.flatten(order='C'))
    
    total = []
    counts = []
    
    if greater:
        for pixel in pixels:
            count = len(img[img == pixel])
            probability = count / size
            if probability > intensity:
                total.append(pixel*count)
                counts.append(count)
        if len(counts) == 0:
            return 0
        return np.sum(total) / np.sum(counts)
    else:
        for pixel in pixels:
            probability = len(img[img == pixel]) / size
            if probability <= intensity:
                total.append(pixel)
        return np.mean(total)
```

### Basic Thresholding

```python
def basic_threshold(img, n=100):
    pixels = np.unique(img.flatten(order='C'))
    threshold = (np.max(pixels) - np.min(pixels)) / 2
    for i in range(n):
        mean1 = np.mean(img[img > threshold])
        mean2 = np.mean(img[img <= threshold])
        threshold = (mean1 + mean2) / 2
    return threshold
```

### Otsu threshold

```python
def otsu_threshold(img):
    pixels = np.unique(img.flatten(order='C'))
    size = len(img.flatten(order='C'))
    
    probabilities = []
    for pixel in pixels:
        probabilities.append(len(img[img == pixel]) / size)
    threshold = 0
    max_variance = -1
    
    for i, intensity in enumerate(probabilities):
        prob1 = pixel_probability(img, intensity, True)
        prob2 = pixel_probability(img, intensity, False)
        mean1 = pixel_mean(img, intensity, True)
        mean2 = pixel_mean(img, intensity, False)
        variance = prob1 * prob2 * (mean1 - mean2)**2
        if max_variance < variance:
            max_variance = variance;
            threshold = pixels[i]
    
    return threshold
```

```python
display(lena_gray, basic_threshold(lena_gray, 100))
```

```python
display(lena_gray, otsu_threshold(lena_gray))
```
