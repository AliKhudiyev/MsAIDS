# Edge detection

```python
cones_gray = cv2.cvtColor(cones, cv2.COLOR_BGR2GRAY)

edges_x = cv2.Sobel(cones_gray, cv2.CV_64F, 1, 0, ksize=3)
edges_y = cv2.Sobel(cones_gray, cv2.CV_64F, 0, 1, ksize=3)

fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].imshow(edges_x, 'gray');
axs[1].imshow(edges_y, 'gray');
```

```python
freqs = []
unique_colors = np.unique(cones_gray)

for color in unique_colors:
    freqs.append(len(cones_gray[cones_gray == color])/(cones_gray.shape[0]*cones_gray.shape[1]))

plt.hist(cones_gray);
```

```python
def get_edges(edges_x, edges_y, threshold=140):   
    edges = edges_x.copy()
    for i in range(edges_x.shape[0]):
        for j in range(edges_x.shape[1]):
            if np.sqrt(edges_x[i][j]**2 + edges_y[i][j]**2) > threshold:
                edges[i][j] = 255
            else:
                edges[i][j] = 0
    return edges

edges = get_edges(edges_x, edges_y)
plt.imshow(edges, 'gray');
```

```python
cones_detected = cones.copy()
for i in range(cones.shape[0]):
    for j in range(cones.shape[1]):
        if edges[i][j] == 255:
            cones_detected[i][j] = [255, 0, 0]

plt.imshow(cones_detected);
```

```python
cones_red_edges_x = cv2.Sobel(cones[:,:,0], cv2.CV_64F, 1, 0, ksize=3)
cones_red_edges_y = cv2.Sobel(cones[:,:,0] ,cv2.CV_64F, 0, 1, ksize=3)
cones_red_edges = get_edges(cones_red_edges_x, cones_red_edges_y)

cones_green_edges_x = cv2.Sobel(cones[:,:,1], cv2.CV_64F, 1, 0, ksize=3)
cones_green_edges_y = cv2.Sobel(cones[:,:,1], cv2.CV_64F, 0, 1, ksize=3)
cones_green_edges = get_edges(cones_green_edges_x, cones_green_edges_y)

cones_blue_edges_x = cv2.Sobel(cones[:,:,2], cv2.CV_64F, 1, 0, ksize=3)
cones_blue_edges_y = cv2.Sobel(cones[:,:,2], cv2.CV_64F, 0, 1, ksize=3)
cones_blue_edges = get_edges(cones_blue_edges_x, cones_blue_edges_y)

fig, axs = plt.subplots(1, 3, figsize=(18, 6))
axs[0].imshow(cones_red_edges, 'gray');
axs[1].imshow(cones_green_edges, 'gray');
axs[2].imshow(cones_blue_edges, 'gray');
```

```python
cones_rgb_edges = cones_red_edges.copy()
for i in range(cones.shape[0]):
    for j in range(cones.shape[1]):
        cones_rgb_edges[i][j] = (cones_red_edges[i][j]+cones_green_edges[i][j]+cones_blue_edges[i][j])//3

cones_edges = cones.copy()
for i in range(cones.shape[0]):
    for j in range(cones.shape[1]):
        if cones_rgb_edges[i][j] > 0:
            cones_edges[i][j] = [255, 0, 0]

plt.imshow(cones_edges, 'gray');
```
