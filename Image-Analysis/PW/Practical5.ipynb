# 1st order statistical methods

```python
random1 = cv2.imread('base-images/random1.png')
random2 = cv2.imread('base-images/random1.png')
pixnio = cv2.imread('base-images/PIXNIO.jpg')
plt.imshow(random2);
```

```python
def gray_levels(img):
    return np.unique(img.flatten(order='C'))

def num_gray_levels(img):
    return len(gray_levels(img))

def probability_pixel(img, p):
    return np.count_nonzero(img == p) / len(img.flatten(order='C'))

def texture_mean(img):
    mean_ = 0
    for pixel in gray_levels(img):
        mean_ += pixel * probability_pixel(img, pixel)
    return mean_

def texture_variance(img, k=2):
    variance_ = 0
    mean_ = texture_mean(img)
    for pixel in gray_levels(img):
        variance_ += (pixel - mean_)**k * probability_pixel(img, pixel)
    return variance_
```

```python
print('------- Random1 -------')
print('mean:', texture_mean(random1))
print('variance:', texture_variance(random1, k=2))
print('skewness:', texture_variance(random1, k=3))
print('Kurtosis:', texture_variance(random1, k=4))

print('\n------- Random2 -------')
print('mean:', texture_mean(random2))
print('variance:', texture_variance(random2, k=2))
print('skewness:', texture_variance(random2, k=3))
print('Kurtosis:', texture_variance(random2, k=4))

print('\n------- PIXNIO -------')
print('mean:', texture_mean(pixnio))
print('variance:', texture_variance(pixnio, k=2))
print('skewness:', texture_variance(pixnio, k=3))
print('Kurtosis:', texture_variance(pixnio, k=4))
```
