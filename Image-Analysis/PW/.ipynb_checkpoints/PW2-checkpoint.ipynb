# 1. 1D case
## 1.1. Ramp and roof edges

```python
def ramp():
    X = range(0, 10)
    Y = [0 if i<5 else 1 for i in range(10)]
    return X, Y

def roof():
    X = range(0, 10)
    Y = [5 if i==5 else 0 for i in range(10)]
    return X, Y
```

```python
X_ramp, Y_ramp = ramp()
X_roof, Y_roof = roof()

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].set_title('Ramp')
axs[1].set_title('Roof')

axs[0].plot(X_ramp, Y_ramp, '.-');
axs[1].plot(X_roof, Y_roof, '.-');
```

## 1.2. Derivatives

```python
Y_ramp_derivative = np.diff(Y_ramp)
Y_roof_derivative = np.diff(Y_roof)

Y_ramp_derivative2 = np.diff(Y_ramp, n=2)
Y_roof_derivative2 = np.diff(Y_roof, n=2)

fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].set_title('Ramp 1st order derivative')
axs[0, 1].set_title('Ramp 2nd order derivative')
axs[1, 0].set_title('Roof 1st order derivative')
axs[1, 1].set_title('Roof 2nd order derivative')

axs[0, 0].plot(range(len(Y_ramp_derivative)), Y_ramp_derivative);
axs[0, 1].plot(range(len(Y_roof_derivative)), Y_roof_derivative);
axs[1, 0].plot(range(len(Y_ramp_derivative2)), Y_ramp_derivative2);
axs[1, 1].plot(range(len(Y_roof_derivative2)), Y_roof_derivative2);
```

# 2. First order derivative operators
## 2.1. Norm of the gradient

```python
def gradient_norm(img):
    gradients = np.gradient(img)
    x_gradient, y_gradient = gradients[0], gradients[1]
    gradient_norm = np.zeros(shape=(gradients.shape[0], gradients.shape[1]))
    for i in range(len(x_gradient)):
        gradient_norm[i][j] = (x_gradient[i]**2 + y_gradient[i]**2)**(1/2)
    return gradient_norm
```

# 4. Thresholding

```python
def binarization(img, threshold=None):
    img_copy = np.copy(img)
    for i in range(img_copy.shape[0]):
        for j in range(img_copy.shape[1]):
            if (img[i][j] - np.min(img)) / (np.max(img) - np.min(img)) < threshold:
                img_copy[i][j] = 0
            else:
                img_copy[i][j] = 1
    return img_copy
```
