## II - Application on real images

```python
def hist(imgs):
    fig, axs = plt.subplots(1,len(imgs),figsize=(6*len(imgs),6))
    
    for i, img in enumerate(imgs):
        pixels = img.flatten(order='C')
        unique_pixels = np.unique(pixels)
        frequencies = np.zeros(shape=(len(unique_pixels)))
        for pixel in pixels:
            frequencies[np.where(unique_pixels == pixel)[0]] += 1
        if len(imgs) > 1:
            axs[i].plot(unique_pixels, frequencies/len(pixels));
        else:
            axs.plot(unique_pixels, frequencies/len(pixels));

def imshow(imgs):
    fig, axs = plt.subplots(1,len(imgs),figsize=(6*len(imgs),6))
    
    for i, img in enumerate(imgs):
        if len(imgs) > 1:
            axs[i].imshow(img, 'gray');
        else:
            axs.imshow(img, 'gray');
```

```python
def cumulated_hist(imgs):
    fig, axs = plt.subplots(1,len(imgs),figsize=(6*len(imgs),6))
    
    for i, img in enumerate(imgs):
        pixels = img.flatten(order='C')
        unique_pixels = np.unique(pixels)
        frequencies = np.zeros(shape=(len(unique_pixels)))
        for pixel in pixels:
            frequencies[np.where(unique_pixels == pixel)[0]] += 1
        for j in range(1,len(frequencies)):
            frequencies[j] += frequencies[j-1];
        if len(imgs) > 1:
            axs[i].plot(unique_pixels, frequencies/len(pixels));
        else:
            axs.plot(unique_pixels, frequencies/len(pixels));
```

```python
cumulated_hist([spine, rice, cameraman])
```

### 2. Linear stretching

```python
def recdyn(img,a,b):
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            img[i][j] = a*img[i][j]+b
```

```python
spine_stretched = copy.deepcopy(spine)
recdyn(spine_stretched,50,140)

rice_stretched = copy.deepcopy(rice)
recdyn(rice_stretched,50,140)

cameraman_stretched = copy.deepcopy(cameraman)
recdyn(cameraman_stretched,50,140)

imshow([spine, rice, cameraman])
hist([spine, rice, cameraman])
imshow([spine_stretched, rice_stretched, cameraman_stretched])
hist([spine_stretched, rice_stretched, cameraman_stretched])
```

### 3. Negative of an image

```python
def negim(img):
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            img[i][j] = 255 - img[i][j]
```

```python
spine_negative = copy.deepcopy(spine)
negim(spine_negative)

rice_negative = copy.deepcopy(rice)
negim(rice_negative)

cameraman_negative = copy.deepcopy(cameraman)
negim(cameraman_negative)

imshow([spine, rice, cameraman])
hist([spine, rice, cameraman])
imshow([spine_negative, rice_negative, cameraman_negative])
hist([spine_negative, rice_negative, cameraman_negative])
```
