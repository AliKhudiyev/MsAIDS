# Forecasting

```python
air_passengers = pd.read_csv('../AirPassengers.csv', header=0, parse_dates=[0], names=['Month', '#Passengers'], index_col=0)
air_passengers.head()
```

```python
log_air_passengers = np.log(air_passengers)
log_air_passengers[:5]
```

```python
log_air_passengers_diff = log_air_passengers - log_air_passengers.shift()
log_air_passengers_diff.dropna(inplace=True)
log_air_passengers_diff.head()
```

```python
from statsmodels.tsa.stattools import acf, pacf

lag_acf=acf(log_air_passengers_diff.values, nlags=20)
lag_pacf=pacf(log_air_passengers_diff.values, nlags=20)

print(lag_acf)
print(lag_pacf)
```

```python
plt.plot(lag_acf);
plt.plot(lag_pacf);
plt.axhline(y=0, linestyle='--');
plt.axhline(y=1.96/np.sqrt(len(log_air_passengers_diff)), linestyle='--');
plt.axhline(y=-1.96/np.sqrt(len(log_air_passengers_diff)), linestyle='--');
```

```python
from statsmodels.tsa.arima_model import ARIMA

# AR model
model = ARIMA(log_air_passengers, order=(2,1,0))
results_predicted_AR = model.fit(disp=-1)

plt.plot(log_air_passengers_diff);
plt.plot(results_predicted_AR.fittedvalues, c='red');
# plt.title('RSS: %.4f' %sum((results_predicted_AR.fittedvalues-log_air_passengers_diff)**2))
```

```python
# MA model
model = ARIMA(log_air_passengers, order=(0,1,2))
results_predicted_MA = model.fit(disp=-1)

plt.plot(log_air_passengers_diff);
plt.plot(results_predicted_MA.fittedvalues, c='red');
# plt.title('RSS: %.4f' %sum((results_predicted_MA.fittedvalues-log_air_passengers_diff)**2))
```

```python
# ARIMA model
model = ARIMA(log_air_passengers, order=(2,1,2))
results_predicted_ARIMA = model.fit(disp=-1)

plt.plot(log_air_passengers_diff);
plt.plot(results_predicted_ARIMA.fittedvalues, c='red');
# plt.title('RSS: %.4f' %sum((results_predicted_ARIMA.fittedvalues-log_air_passengers_diff)**2))
```

```python
pred_ARIMA_diff = pd.Series(results_predicted_ARIMA.fittedvalues, copy=True)
pred_ARIMA_diff.head()
```

```python
pred_ARIMA_diff_cumsum = pred_ARIMA_diff.cumsum()
pred_ARIMA_diff_cumsum.head()
```

```python
pred_ARIMA_log = pd.Series(log_air_passengers['#Passengers'], index=log_air_passengers.index)
pred_ARIMA_log = pred_ARIMA_log.add(pred_ARIMA_diff_cumsum, fill_value=0)
pred_ARIMA_log.head()
```

```python
plt.plot(log_air_passengers);
plt.plot(pred_ARIMA_log);
```
